```python
import tkinter as tk
from tkinter import ttk

def sigmoid(x):
    return 1 / (1 + (2.718281828459045 ** -x))

def sigmoid_derivative(x):
    return x * (1 - x)

def update_weights():
    input1 = float(entry_input1.get())
    input2 = float(entry_input2.get())
    target_o1 = float(entry_target_o1.get())
    target_o2 = float(entry_target_o2.get())
    eta = 0.5

    w5, w6, w7, w8 = 0.40, 0.45, 0.50, 0.55
    b2 = 0.60

    net_h1 = 0.15 * input1 + 0.20 * input2 + 0.35
    out_h1 = sigmoid(net_h1)

    net_h2 = 0.25 * input1 + 0.30 * input2 + 0.35
    out_h2 = sigmoid(net_h2)

    net_o1 = w5 * out_h1 + w6 * out_h2 + b2
    out_o1 = sigmoid(net_o1)

    net_o2 = w7 * out_h1 + w8 * out_h2 + b2
    out_o2 = sigmoid(net_o2)

    d_E_total_out_o1 = -(target_o1 - out_o1)
    d_out_o1_net_o1 = sigmoid_derivative(out_o1)
    d_net_o1_w6 = out_h2
    d_E_total_w6 = d_E_total_out_o1 * d_out_o1_net_o1 * d_net_o1_w6
    w6_new = w6 - eta * d_E_total_w6

    d_E_total_out_o2 = -(target_o2 - out_o2)
    d_out_o2_net_o2 = sigmoid_derivative(out_o2)
    d_net_o2_w7 = out_h1
    d_E_total_w7 = d_E_total_out_o2 * d_out_o2_net_o2 * d_net_o2_w7
    w7_new = w7 - eta * d_E_total_w7

    d_net_o2_w8 = out_h2
    d_E_total_w8 = d_E_total_out_o2 * d_out_o2_net_o2 * d_net_o2_w8
    w8_new = w8 - eta * d_E_total_w8

    label_w6_new.config(text=f"w6_new: {w6_new:.5f}")
    label_w7_new.config(text=f"w7_new: {w7_new:.5f}")
    label_w8_new.config(text=f"w8_new: {w8_new:.5f}")

root = tk.Tk()
root.title("Neural Network GUI")
root.geometry("400x400")
root.configure(bg="#2c3e50")

style = ttk.Style()
style.configure("TButton", font=("Arial", 12), padding=5)
style.configure("TLabel", font=("Arial", 12), background="#2c3e50", foreground="white")
style.configure("TEntry", font=("Arial", 12))

entry_input1 = ttk.Entry(root)
entry_input1.insert(0, "0.05")
entry_input2 = ttk.Entry(root)
entry_input2.insert(0, "0.10")
entry_target_o1 = ttk.Entry(root)
entry_target_o1.insert(0, "0.01")
entry_target_o2 = ttk.Entry(root)
entry_target_o2.insert(0, "0.99")

btn_calculate = ttk.Button(root, text="Update Weights", command=update_weights)

label_w6_new = ttk.Label(root, text="w6_new: ")
label_w7_new = ttk.Label(root, text="w7_new: ")
label_w8_new = ttk.Label(root, text="w8_new: ")

ttk.Label(root, text="Input1: ").grid(row=0, column=0, pady=5, padx=5)
entry_input1.grid(row=0, column=1, pady=5, padx=5)

ttk.Label(root, text="Input2: ").grid(row=1, column=0, pady=5, padx=5)
entry_input2.grid(row=1, column=1, pady=5, padx=5)

ttk.Label(root, text="Target O1: ").grid(row=2, column=0, pady=5, padx=5)
entry_target_o1.grid(row=2, column=1, pady=5, padx=5)

ttk.Label(root, text="Target O2: ").grid(row=3, column=0, pady=5, padx=5)
entry_target_o2.grid(row=3, column=1, pady=5, padx=5)

btn_calculate.grid(row=4, column=0, columnspan=2, pady=10)

label_w6_new.grid(row=5, column=0, columnspan=2, pady=5)
label_w7_new.grid(row=6, column=0, columnspan=2, pady=5)
label_w8_new.grid(row=7, column=0, columnspan=2, pady=5)

root.mainloop()
```
